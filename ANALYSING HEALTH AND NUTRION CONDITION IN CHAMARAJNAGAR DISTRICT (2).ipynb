# BARCHART REPRESENTING THE IMMUNIZATION OF CHILDREN FOR DPT,POLIO,BCG 

```python
import matplotlib.pyplot as plt
import numpy as np
x= [2015,2016,2017,2018,2019]
DPT = [11066, 11658, 12157, 12458, 13152]
POLIO = [11658, 12117, 12458, 13568, 14152]
BCG = [11058, 11986, 11856, 12798, 13745]

b_BCG =list(np.add(DPT,POLIO))

plt.bar(x, DPT, 0.4,label='DPT' )
plt.bar(x, POLIO, 0.4,bottom = DPT,label='POLIO' )
plt.bar(x, BCG, 0.4,bottom =b_BCG, label='BCG' )
plt.yticks([0,10000,20000,30000,40000],['0','10k','20k','30k','40k'])

plt.xlabel('YEARS')
plt.ylabel('TOTAL NUMBER')
plt.title('TOTAL IMMUNIZATION ')
plt.legend(loc=(1,1))

plt.savefig('IMMUNIZE.png')
```

# GRAPH SHOWING DOCTORS RATIO IN EACH TALUK CATEGORISED INTO LOW(<=4 DOCTORS/LAC),MEDIUM(5-10 DOCTORS/LAC) & HIGH (>10 DOCTORS/LAC) AS PER IPHS GUIDLINES 

```python
import numpy as np
import matplotlib.pyplot as plt
# set width of bar
barWidth = 0.15
LOW = [9,2,9,3]
MEDIUM= [6,11,9,2]
HIGH= [5,6,2,0]
# Set position of bar on X axis
r1 = np.arange(len(LOW))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3] 
# Make the plot
plt.bar(r1, LOW, color='r', width=barWidth, edgecolor='white', label='LOW')
plt.bar(r2, MEDIUM, color='y', width=barWidth, edgecolor='white', label='MEDIUM')
plt.bar(r3, HIGH, color='b', width=barWidth, edgecolor='white', label='HIGH')
# Add xticks on the middle of the group bars
plt.xlabel('TALUK', fontweight='bold')
plt.ylabel(' NO OF PHC', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(LOW))], ['CHAMARAJNAGAR', 'GUNDLUPET', 'KOLLEGAL', 'YELANDUR'])
plt.title('DOCTORS RATIO')
# Create legend & Show graphic
plt.legend(loc=(1,0))
plt.show()
plt.savefig('DOCTOR.png')
```
